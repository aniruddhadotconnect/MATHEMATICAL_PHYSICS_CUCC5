```
import matplotlib.pyplot as plt
%matplotlib inline

X, Y = [], []
for line in open('/content/sample2.txt', 'r'):
  values = [int(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])


for i in range(len(X)):
    print(X[i],Y[i])


plt.plot(X,Y,'o-')
plt.show()
```
