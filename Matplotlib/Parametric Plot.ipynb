```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
#####################  Plot of a Circle  ################
r = 4 # Radius
wt = np.linspace(0,2*np.pi,20)
x = r*np.cos(wt)
y = r*np.sin(wt)  # Putting 2r will make it ellipse
plt.plot(x,y)
plt.grid()
plt.axes().set_aspect('equal')
plt.show() 
```

# Lissajous Figure

```
t = np.linspace(0,1,100)
wt = 2*np.pi*t
plt.plot(np.sin(wt + (3*np.pi/4)),np.cos(wt))
plt.axes().set_aspect('equal')
plt.grid()
plt.show()
```

```
t = np.linspace(0,1,100)
wt = 2*np.pi*t
plt.plot(np.sin(4*wt),np.cos(wt))
plt.axes().set_aspect('equal')
plt.grid()
plt.show()
```

# Fourier Analysis

```
wt = np.linspace(0,2*np.pi,100)
y = np.sin(wt)
for n in range(3,10,2):
    y += np.sin(n*wt)/n   #Square wave
plt.plot(wt,y)
plt.show()
```
