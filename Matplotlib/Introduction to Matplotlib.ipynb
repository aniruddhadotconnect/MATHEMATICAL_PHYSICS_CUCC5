Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. Pyplot is a Matplotlib module which provides a MATLAB-like interface. Actual author id John D Hunter. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python, and the advantage of being free and open-source. 

```
import numpy as np
# This Numpy inherently works with MATPLOTLIB

import matplotlib.pyplot as plt
#%matplotlib inline    ## MAGIC COMMAND: It helps to view the plot within working environment
```

```
x = np.linspace(0,10,20)
y = x



plt.plot(x,y)
plt.show()                  
```

###Adding GRID to plot
Insert plot.grid()    or    plt.grid(True)   before plt.show()

```
y = x**2



plt.plot(x,y)
plt.grid(True)           ## This command helps to add the grid
plt.show()
```

### Adding TITLE to the plot

```
y = x**3



plt.plot(x,y)
plt.title('Demo plot')               ## This adds title to plot
plt.grid(True)
plt.show()
```

### Adding LABELs to the axes
*   plt.xlabel('x name')
*   plt.xlabel('y name')




```
y = x**3



plt.plot(x,y)
plt.xlabel('x range of real numbers')        ## Adds label to X axis
plt.ylabel('y values')                       ## Adds label to Y axis
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

### Limiting the Axis Values

*   plt.xlim([a,b])
*   plt.ylim([c,d])



```
y = x**3



plt.plot(x,y)
plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.xlim([0,8])                            ## puts a limit on X axis 
plt.ylim([0,600])                          ## puts a limit on Y axis, here origin(0,0)
plt.show()
```

### Subplots

###Multiline Plots
Multiple functions in a single plot.Automatically makes the color of each plot different

```
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.plot(x,2**x)


plt.xlabel(' range of real numbers')      
plt.ylabel('Functions')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

```
plt.plot(x,x**2,x,x**3,x,2**x)        ####   Multiline plot in a single command


plt.xlabel(' range of real numbers')      
plt.ylabel('Functions')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

### Adding Legend
It's barely possible to understand that which plot belongs to which function just by looking at the plot.That's why we use legend.:    plt.legend()

```
plt.plot(x,x**2, label = 'x**2')              ## label is added to each plot
plt.plot(x,x**3, label = 'x**3')
plt.plot(x,2**x, label = '2**x')
plt.legend()


plt.xlabel(' range of real numbers')      
plt.ylabel('Functions')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

```
plt.plot(x,x**2,x,x**3,x,2**x)        
plt.legend(['x**2','x**3','2**x'])                  ## Legend with different syntax


plt.xlabel(' range of real numbers')      
plt.ylabel('Functions')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

####Legend Position:
Location of the legend can be changed ocer the plot.
*  plt.legend(loc = 'upper center')
*  plt.legend(loc = 'center' )


```
plt.plot(x,x**2,x,x**3,x,2**x)        
plt.legend(['x**2','x**3','2**x'], loc = 'upper center')                  ## Legend with different syntax


plt.xlabel(' range of real numbers')      
plt.ylabel('Functions')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

### Plot Color
These are available colors 
*   green : g
*   Blue : b
*   Yellow : y

*   Black : k

```
plt.plot(x,x**2,'g')     ## green



plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

###Plot Markers
Markers are just the data points. Only they represents SCATTER plots.
*   'o' :  Circle
*   ''<'' : Triangle 1
*   ''>'' : Triangle 2
*   'v' : Triangle 3
*   ''*'' : Star
*   '+' : Plus

there are other markers also



```
plt.plot(x,x**2,'v')             ## with triangular markers
plt.plot(x,x**2.5,'+')           ## with plus markers



plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

##Line Styles


1.   '-' : Solid Line
2.   '- -' : Dashed line
3.   ' : ' : dotted line
4.   ' -.' : Dash-dot line



```
plt.plot(x,x,'-')
plt.plot(x,x*1.8,'--')
plt.plot(x,x**1.5,'-.')
plt.plot(x,x**2,':')
plt.legend(['x','x**2','x**3','2**x'], loc = 'upper center')    



plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

### General Color-Marker-Style Command
plt.plot(x,y,'color Marker Style')

```
plt.plot(x,x,'r*--')



plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

```
plt.plot(x,x,'r*-')                             ## Multiline plots
plt.plot(x,x*1.8,'go--')
plt.plot(x,x**1.5,'k<-.')
plt.plot(x,x**2,'m+:')
plt.legend(['x','x**1.8','x**1.5','x**2'], loc = 'center')    


plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

```
### Multiline with CMS plot in one command
plt.plot(x,x,'r*-',x,x*1.8,'go--',x,x**1.5,'k<-.',x,x**2,'m+:')
plt.legend(['x','x**1.8','x**1.5','x**2'], loc = 'upper left')

plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

##FINAL CUSTOMIZATION


*   color : r,g,y,b,k,m,w
*   linestyle : '- -',  '-',   '-.',   ':'
*   linewidth : positive Real number
*   marker : o,<,>,+,*,v
*   markerfacecolor : r,g,y,b,k,m,w
*   markeredgecolor : r,g,y,b,k,m,w
*   markersize : positive real number


```
plt.plot(x,x**2,
         color='r',
         linestyle='--',
         linewidth=20,
         marker='<',
         markerfacecolor='y',
         markeredgecolor='k',
         markeredgewidth=2,
         markersize=10)


plt.xlabel(' range of real numbers')      
plt.ylabel('y values')                      
plt.title('Demo plot')              
plt.grid(True)
plt.show()
```

# More types of plots

### Histogram

```
z = np.random.randint(0,20,100)
print(z)

plt.hist(z)
plt.show()
```

```
z = np.random.randint(0,20,100)
print(z)

plt.hist(z,bins= 20,facecolor = 'b', edgecolor = 'k', alpha = 0.58)  # alpha = transparency
plt.grid()
plt.show()
```

```
z = np.random.normal(50,2,1000)  ## normal distribution (mean,std deviation, number of elements)

plt.hist(z, 80, edgecolor = 'k')  # bins = 80
plt.show()
```

### Scatter plot

```
z = np.random.normal(50,2,1000)  ## normal distribution (mean,std deviation, number of elements)
x = np.linspace(-100,100,1000)
plt.scatter(z,x)  
plt.show()
```

```
z = np.random.normal(50,2,1000)  ## normal distribution (mean,std deviation, number of elements)
x = np.linspace(-100,100,1000)


plt.scatter(z,x, facecolor = 'r', edgecolor = 'k', alpha = 0.8)  
plt.show()
```

```
z = np.random.normal(50,2,100)  ## normal distribution (mean,std deviation, number of elements)
x = np.linspace(-100,100,100)


plt.scatter(z,x, facecolor = 'r', edgecolor = 'k', alpha = 0.8,
            s = 700*np.random.random(100), c = np.random.random(100))  
plt.show()
```

### Pie chart

```
x = [10,20,30,40]
label = ['C++','C','Fortran','Python']

## we cannot use dictionary

plt.pie(x,labels = label)
plt.show()
```

```
x = [10,20,30,40]
label = ['C++','C','Fortran','Python']
explode = [0.12,0.15,0.01,0.1]

## we cannot use dictionary

plt.pie(x,labels = label,explode = explode)
plt.show()
```

### Meshgrid

```
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
X,Y = np.meshgrid(x,y)
plt.plot(X,Y,'.',ms =20)
plt.show()
```

### Contour plot

```
Z = np.sin(X)**2 + np.cos(Y)**2
plt.contour(X,Y,Z)
plt.show()
```

```
Z = np.sin(X)**2 + np.cos(Y)**3
plt.contour(X,Y,Z, colors = 'k')
plt.show()
```

```
## contour filled

Z = np.sin(X)**2 + np.cos(Y)**2
plt.contourf(X,Y,Z)
plt.show()
```

```
## contour filled with color  map

Z = np.sin(X)**2 + np.cos(Y)**2
plt.contourf(X,Y,Z,500,cmap = 'hot')   ## here 500 is edge blur index
plt.show()
```

```
## contour filled with color  map

Z = np.sin(X)**2 + np.cos(Y)**2
plt.contourf(X,Y,Z, cmap = 'binary')
plt.show()
```

```
## contour filled with color  map

Z = np.sin(X)**4 + np.cos(Y)**4
plt.contourf(X,Y,Z,cmap = 'cool')
plt.show()
```
