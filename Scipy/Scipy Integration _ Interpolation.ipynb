# Quad Integration

```
# integrate using quad
from scipy import integrate
import numpy as np


def f(x):
    return np.sin(x)

result = integrate.quad(f,0,np.pi)
print(result)   #--------------------Answer is (result, error)
```

# Interpolation 

```
###################################   Problem 1
from scipy import interpolate

X = np.array([1,3,4,6,7,8])
Y = np.array([10,34,39,60,73,81])

f = interpolate.interp1d(X,Y) #--------------- kind = 'linear'

f(5.4) #-----------------------------resulf of the 1d interpolation 
```

```
## Scipy interpolate example 2

import matplotlib.pyplot as plt
%matplotlib inline
from scipy import interpolate

X = np.arange(0,10,1.5)
Y = np.exp(-X)
f = interpolate.interp1d(X,Y)

X_test = np.linspace(1,9,4)
Y_test = f(X_test)
#print(Y_test)


plt.plot(X,Y,'o-',X_test,Y_test,'>g')
plt.show()
```
