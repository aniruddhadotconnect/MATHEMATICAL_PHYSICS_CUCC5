## Solving Differential Equation using Python

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```
def f(y,x):
    k = 0.4
    dydx = -k*y
    return dydx

x = np.linspace(0,10,20)
y0 = 100 #------------------------- Initial Condition
y = odeint(f,y0,x)



plt.plot(x,y,'o-')
plt.grid(which = 'both',linestyle = '-',alpha = 0.4)
plt.title('Decay Curve',fontsize = 20)
plt.xlabel('x', fontsize = 20)
plt.ylabel('f(x)', fontsize = 20)
plt.show()
```

### Taking 'k' as parameter 

```
def f(y,x,k):
    dydx = -k*y
    return dydx

x = np.linspace(0,10,20)
y0 = 100 #------------------------- Initial Condition


y1 = odeint(f,y0,x, args = (0.2,))
y2 = odeint(f,y0,x, args = (0.5,))
y3 = odeint(f,y0,x, args = (0.8,))

plt.plot(x,y1,x,y2,x,y3)
plt.legend(['k=0.2','k=0.5','k=0.8'])
plt.grid(which = 'both',linestyle = '-',alpha = 0.4)
plt.title('Decay Curve',fontsize = 20)
plt.xlabel('x', fontsize = 20)
plt.ylabel('f(x)', fontsize = 20)
plt.show()
```

# 2nd order ODE using Scipy

```
##### 2nd Order ODE for Damped Harmonic Oscillator ############
k1,k2 = 1,0.15

def f(u,t):
    x = u[0]
    y = u[1]
    dxdt = y
    dydt = -(k1*x) -(k2*y)
    return (np.array([dxdt,dydt]))

u0 = [1,0]
t = np.linspace(0,30,60)
solution = odeint(f,u0,t)

X = solution[:,0]
Y = solution[:,1]

plt.plot(t,X)
plt.grid()
plt.show()
```

```
###########  Forced oscillation ##############

k1,k2 = 1,0.15

def f(u,t):
    x = u[0]
    y = u[1]
    dxdt = y
    dydt = -(k1*x) -(k2*y) + np.cos(t)
    return (np.array([dxdt,dydt]))

u0 = [1,0]
t = np.linspace(0,100,500)
solution = odeint(f,u0,t)

X = solution[:,0]
Y = solution[:,1]

plt.plot(t,X)
plt.grid()
plt.show()
```

# Lorentz Attractor: Coupled Differential Equation using Scipy

```
################  Lorentz Attractor  ###################
sigma, rho, beta = 10,28,8/3

def f(u,t):
    x,y,z = u[0],u[1],u[2]
    dxdt = sigma*(y - x)
    dydt = x*(rho - z) - y
    dzdt = (x*y) - (beta*z)
    return ([dxdt,dydt,dzdt])

u0 = [0,1.0,0]
t = np.linspace(0,30,1000)

sol = odeint(f,u0,t)

X,Y,Z = sol[:,0],sol[:,1],sol[:,2]
########################################################
plt.subplot(221)
plt.plot(Y,Z)
plt.subplot(222)
plt.plot(X,Z)
plt.subplot(223)
plt.plot(X,Y)
plt.subplot(224)
plt.plot(t,X,t,Y,t,Z)
plt.show()
```
