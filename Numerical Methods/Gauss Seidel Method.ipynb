Solve thses linear equations

5$x_1$ - $x_2$ + 2$x_3$ = 12 \\
3$x_1$ + 8$x_2$ - 2$x_3$ = -25 \\
$x_1$ + $x_2$ + $x_3$ = 6


```
#########################  GAUSS - SEIDEL METHOD 1  ############################


x1,x2,x3 = 0,0,0     #----------- Guess solution
print("  X1                              X2                               X3")

for _ in range(10):
    x1 = (x2 - 2*x3 + 12)/5
    x2 = (2*x3 - 3*x1 - 25)/8
    x3 = (6 - x1 - x2)
    print(f"{round(x1,3):->{5}}  {round(x2,3):->{30}}  {round(x3,3):->{30}}")


print('\n \n \n Final solution is: ',x1,x2,x3)
```
