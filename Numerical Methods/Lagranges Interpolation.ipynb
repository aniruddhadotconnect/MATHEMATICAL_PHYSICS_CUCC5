```
#############    GIVEN DATA FROM QUESTION     ################

n = int(input("Enter the number of given data points: "))
X=[]
Y=[]
for i in range(n):
    X.append(float(input("Enter X value: ")))
    Y.append(float(input("Enter corresponding Y value: ")))

print(" X: ",X,"\n","Y: ",Y)


k = float(input("Point for the interpolation: "))
```

```
def lagranges_interpolation(k,X,Y):
    L=[]
    value= 0
    for x in X:
        t=1
        for i in range(n):
            if x == X[i]:
                i+=1
            else:
                t= t*((k - X[i])/(x - X[i]))
        L.append(t)        
#    print(L)

    for i in range(n):
        value += L[i]*Y[i]

    return value

```

```
print("Answer for the Lagranges interpolation is: ",lagranges_interpolation(k,X,Y))
```
