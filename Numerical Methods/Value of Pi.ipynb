```
import numpy as np
import matplotlib.pyplot as plt

n = int(input('Enter the number of: '))
counts = 0
X,Y = [],[]

for i in range(n):
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)
    if (x*x) + (y*y) <= 1:
        X.append(x)
        Y.append(y)
        counts = counts + 1
    
print('value of pi: ', 4*counts/n)
```

```
wt = np.linspace(0,2*np.pi,50)
#X1 = np.cos(wt)
#Y1 = np.sin(wt)
plt.plot(np.cos(wt),np.sin(wt))
plt.scatter(X,Y)

plt.xlim(-1,1)
plt.ylim(-1,1)
plt.axes().set_aspect('equal')
plt.grid()
plt.show()
```
