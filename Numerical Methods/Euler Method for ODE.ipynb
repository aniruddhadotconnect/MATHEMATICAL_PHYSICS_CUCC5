#1st order ODE by Euler Method
###Decay Curve

```
n = int(input("Number of intervals: "))       # n = 20
h = float(input("Enter the value of h :"))    # h = 0.5


### At x =0, y= 100 is the given initial condition
print('Enter the initial condition')
x = float(input("The x value is :"))
y = float(input("The y value is : "))
k = float(input('Enter decay constant(0 < k <= 1): '))

X,Y = [],[]


print("   x                              y")
for i in range(n):
    print(f"{round(x,3):->{5}}  {round(y,3):->{30}}")
    y = y + h*(-k*y)
    x = x + h

    X.append(x)
    Y.append(y)
    
    


#####################      PLOTTING THE SOLUTION      ####################

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(X,Y)
plt.title('Decay Curve')
plt.show()
```

#2nd Order ODE by Euler Method
###Damped Oscillation

```
n = int(input("Number of intervals: "))     # n = 60
h = float(input("Enter the value of h :"))  # h = 0.5


### At t = 0, x =1, y= 0 is the given initial condition
print('Enter the initial condition')
t = float(input("For the given t:"))
x = float(input("The x value is :"))
y = float(input("The y value is : "))


T,X,Y = [],[],[]


print("   t                              x                                       y")
for i in range(n):
    print(f"{round(t,3):->{5}}  {round(x,3):->{30}}   {round(y,3):->{40}}")

    x = x + h*y
    y = y + h*(- (x) - (0.15*y)) 
    t = t + h
    X.append(x)
    Y.append(y)
    T.append(t)

#####################      PLOTTING THE SOLUTION      ####################

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(T,X)
plt.title('Displacement')
plt.show()
```
