#Bisection Method:

```
def bisection_root(a,b):
    while f(a)*f(b) <0:
        c = (a+b)/2
        
        #print(round(a,3),round(f(a),3),round(b,3),round(f(b),3),round(c,3),round(f(c),3))
        print(f"{a:->{10}}  {b:->{10}}   {c:->{15}}   {f(c):->{40}}")
    
        if f(a)*f(c) <0:
            b=c
            if abs(b-a) < 0.01:
                return c

        elif f(c)*f(b) <0:
            a=c
            if abs(b-a) < 0.01:
                return c

        elif f(c)==0:
            return c
    
```

```
def f(x):                       ## Function 
    return (x*x*x)-x-2


#############################  PROGRAM EXECUTION   ####################################

Run = True
while Run == True:
    a = float(input("Enter the lower limit of the root: "))
    b = float(input("Enter the upper limit of the root: "))
    if f(a)*f(b) <0:
        print('The final root is: ',round(bisection_root(a,b),3))
        Run = False
    
```

#Newton Raphson Method:

```
def f(x):          ############ Function
    return (x**3) - x -2
    
def df(x):         ############ Derivative of the function
    return (3*(x**2)) - 1
    
def h(x):
    return f(x)/df(x)

def newton_raphson_root(x):
    X = []    
    while abs(h(x)) > 0.0001:
        x = x-h(x)        
        print(x)
        X.append(x)      ####### prints the values in vertical coloumn        
    return X
```

```
x = float(input("Gues the root "))
X = newton_raphson_root(x)
```
