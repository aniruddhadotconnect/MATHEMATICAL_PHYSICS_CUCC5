```
import numpy as np
X = [5,10,15,20,25,30]
Y = [45,105,174,259,364,496]

D = [Y]
y = Y
for i in range(len(Y)-1):
    y = np.diff(y)
    D.append(y)

h = np.mean(np.diff(X))
print('h is',h)


############ Creating Forward Difference Table

for d in D:
    print(d)
```

```
def term(t,n):
    coeff = 1
    for i in range(n):
        coeff = coeff * (t - i)
    return coeff/np.math.factorial(n)

x = float(input('Where do wnat to interpolate: '))
t = (x - X[0])/h    
print(t)
```

```
sum = 0
for j in range(len(D)):
    sum = sum + term(t,j)*D[j][0]


print('So value of the function at {} is'.format(x),sum)
```
