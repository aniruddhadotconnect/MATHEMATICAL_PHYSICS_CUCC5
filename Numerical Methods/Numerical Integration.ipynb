#Simpson's 1/3rd Rule

In numerical analysis, Simpson’s 1/3 rule is a method for numerical approximation of definite integrals. 
In order to integrate any function f(x) in the interval (a, b), follow the steps given below:


1.  Select a value for n, which is the number of parts the interval is divided into.
2.   Calculate the width, h = (b-a)/n
3.   Calculate the values of x0 to xn as x0 = a, x1 = x0 + h, …..xn-1 = xn-2 + h, xn = b.
Consider y = f(x). Now find the values of y(y0 to yn) for the corresponding x(x0 to xn) values.

```
import numpy as np


a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))
n = int(input("Enter the number of segments: "))
h = (b-a)/n
X=np.linspace(a,b,n)




print("lower limit: {}, upper limit: {} and each of the {} equal intervals is {}".format(a,b,n,h),'\n\n')
print('The x values are {}'.format(X))
```

```
3## Defining the function
import math

def f(x):
    return x**2
```

```
Y=[]
for x in X:                  ###### Creates a list of y values corresponding to x 
    Y.append(f(x))

for i in range(len(X)):    ###### Printing x and y values in a table
    print(X[i],'\t',Y[i])
```

```
###############      Calculation of the sum of EVEN and ODD terms

sum_even = 0
sum_odd  = 0

i=2
while i < len(X)-1:
    sum_even += Y[i]
    i+=2

j=1
while j < len(X)-1:
    sum_odd += Y[j]
    j+=2
    

print("The sum of the evens is {} and sum of the odds is {}".format(sum_even,sum_odd))
```

```
I= (Y[0]+Y[n-1]+ (4*sum_odd) + (2*sum_even))*h/3



print("Result of the integration using Simpson's 1/3 rule is {} ".format(I))
```
