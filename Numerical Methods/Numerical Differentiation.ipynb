##Forward Dfference Method:

```
import numpy as np

X = np.array([2,4,6,8,10])
Y = np.array([10,12.5,16,21.8,32.4])
D = []
y = Y
for i in range(len(Y)-1):
    y = np.diff(y)
    D.append(y)

h = np.mean(np.diff(X))
print('h is',h)


############ Creating Forward Difference Table
print(Y)
for d in D:
    print(d)
```

```
j = int(input("On which point do you want to calculate the differentiation? "))  ##enter 0
sum = 0

for i in range(len(D[j])):
    sum += ((D[i][0])/(i+1))*((-1)**i)

print((sum)/h)
```
