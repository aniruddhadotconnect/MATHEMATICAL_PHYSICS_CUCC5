# Numpy built-in methods
There are several methods to generate array in numpy arrays.Some of them are :


1.   np.arange()
2.   np.linspace()
3.   np.zeros()
4.   np.ones()
5.   np.identity()



###ARANGE

```
x = np.arange(0,10)     ## np.arange(START,STOP)
print(x)

y = np.arange(5,20,3)   ## np.arange(START,STOP,INTERVAL)
print(y)

# this method does not allow the STOP point. It breaks before STOP
```

#RANDOM ARRAYs
numpy has a lots of ways to generate random number arrays.

###np.random.rand(n)
This creates n random numbers with a distribution between [0,1]. It creates an uniform distribution.


```
np.random.rand(5)               ##  1D array wiith random numbers between 0 to 1
```

```
np.random.rand(4,4)             ##  2D 4x4 array wiith random numbers between 0 to 1
```

###np.random.randn(n)
Return a sample (n elements) from the "standard normal" distribution. Unlike rand which is uniform.

```
np.random.randn(4)                ## follows normal distributions around 0
```

```
np.random.randn(3,3)              ## follows normal distributions around 0
```

###np.random.randint(n)

```
np.random.randint(8)           ## gives a random integer within a range from 0
```

```
np.random.randint(20,25)        ## gives random number within a defined range
```

```
np.random.randint(20,25,10)     ## np.random.randint(START,STOP, NO OF ELEMENTS)
```

#Array Indexing

```
## 1D array indexing

x = np.random.randint(5,50,10)
print(x)                       


print(x[6],'is the 6th element.')      ##This indexing style is similar to list indexing
```

```
## 2D array indexing

x = np.random.randint(0,50,9)
print(x,'\n\n')

y = x.reshape(3,3)
print(y)                       ## It's just a 3x3 aray with random integers
```

```
print(y[1],'is the first row')
print(y[:,2],'is the 2nd column. It is expressed in row format.') #':' means all row elements
print(y[1,2],'is the element of 2nd row and 3rd column. ')
```

#Numpy Operations

```
my_list1 = [2,3,4]       
my_list2 = [5,6,7]


print(2*my_list1,'List multiplication')          ## python list multiplication works as concatination
print(my_list1 + my_list2,'List addiition')      ## addition is also concatination
```

```
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])

print(2*x,'    Numpy 1D array Scalar Multiplication')
print(x+y,'Numpy 1D array addition')
print(y-x,'    Numpy 1D array Subtraction')
print(x*y,'Numpy array Multiplication: Element wise multiplication')
print(x**3,'Numpy cubic operation')
```

###Mathematical Universal Operations

```
print(np.sin(x),'              SinX')
print(np.cos(x),'              CosX')
print(np.sqrt(x),'            Square root')
print(np.log(x),'             Logarithm')
print(np.exp(x),'         Exponential')
```

###Matrix Operations(2D)

```
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])
print(x,'x matrix\n')
print(y,'y matrix')
```

```
x+y      ## Matrix Addition
```

```
x@y      ## Scalar Multiplication(dot product)
```

```
x.dot(y) ## Scalar multiplication(Dot product)    **** not DOT product
```

```
x*y      ## Elementwise multiplication
```

#Statistical Functions

```
a = np.random.randint(0,10,50)
print(a)
```

```
print(np.median(a),'Statistical Median')
print(np.mean(a),'Statistical Mean')
print(np.average(a),'Average')
print(np.std(a),'Standard deviation')
print(np.var(a),'Variance')
```

```
np.histogram(a)
```

#Bitwise Operation

```
a = np.array([0,1,1,0])
b = np.array([0,0,1,1])
print(np.bitwise_and(a,b),'    Bitwise AND')
print(np.bitwise_or(a,b), '    Bitwise OR')
print(np.bitwise_xor(a,b), '    Bitwise XOR')
print(np.bitwise_not(a), 'Bitwise NOT is a unary operation')
```
