```
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```
####  Legendre's polynomial using Rodrigue's formula  ###

n = int(input('Enter the order: '))
X = np.linspace(-1,1,20)
S = np.poly1d([1,0,-1])
Pn = (1/((2**n)*(np.math.factorial(n))))*((S**n).deriv(n))
print('The legendre polynomial is: \n',Pn)
plt.plot(X,Pn(X))
plt.axhline(y = 0)
plt.show()
```

```
############  Bessel Function Plotting using FUNCTION  ##########


def J(n):
    X = np.linspace(0,20,100)
    Y = []
    for x in X:
        sum = 0
        for m in range(0,30):
            sum += (((-1)**m)/(math.factorial(m)*math.factorial(m+n)))*(x/2)**(n+ 2*m)
        Y.append(sum)
    plt.plot(X,Y)
    plt.xlim(0,20)
    plt.ylim(-1,1)
    plt.grid()
    return plt.show()

J(0)

```

```
############  Bessel Function Plotting Analytically  ##########

x = np.linspace(0,30,100)
J0,J1 = 0,0
S = np.poly1d([0,1,0])
for i in range(200):
    J0 += (((-1)**i)*(S**(2*i)))/((2**(2*i))*(np.math.factorial(i)**2))
    J1 +=  (((-1)**i)*(S**(2*i+1)))/(((i+1)*2**(2*i))*(np.math.factorial(i)**2))


plt.plot(x,J0(x),x,J1(x))
plt.legend(['J0','J1'])
plt.grid()
plt.show()
```

```
############  Legendre polynomial Plotting Analytically  ##########


def P(n):
    X = np.linspace(-20,20,100)
    Y = []
    for x in X:
        sum = 0
        for m in range(0,n):
            sum += (((-1)**m)*math.factorial(2*n-2*m)*x**(n-m))/((2**n)*math.factorial(m)*math.factorial(n-m)*math.factorial(n-m))
        Y.append(sum)
    plt.plot(X,Y)
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.grid(color = 'y')
    return plt.show()

P(2)
```
