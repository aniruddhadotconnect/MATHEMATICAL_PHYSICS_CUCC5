# Legendre's Polynomial

```
import scipy.special as s
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<img class="irc_mi" src="https://d2vlcm61l7u1fs.cloudfront.net/media%2Ffff%2Ffff06006-3e23-4790-9012-d24ef2838893%2FphpQ5slwD.png" style="" alt="Image result for legendre differential equation image" width="450" height="144">

<img class="irc_mi" src="https://solitaryroad.com/c679/ole2.gif" style="" alt="Related image" width="253" height="187">

```
print(s.legendre(1))
print(s.legendre(2))
print(s.legendre(3))
```

```
P3 = s.legendre(3)
print(P3[3])    # getting the coefficients
```

```
x = np.arange(-1,1,0.1)
y = P3(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

```
for n in range(5):
    x = np.arange(-1,1,0.1)
    Pn = s.legendre(n)
    y = Pn(x)
    plt.plot(x,y)

plt.grid()
plt.show()
```

```
s.legendre(3)
```

```
s.legendre(3)(2)  # value of legendre polynomial at any point
```

# Hermite Polynomial

<img class="irc_mi" src="https://figkid.files.wordpress.com/2016/09/hermite-polynomial1.png?w=720" style="" alt="Related image" width="397" height="177">

<img class="irc_mi" src="https://solitaryroad.com/c680/ole1.gif" style="" alt="Image result for hermite differential equation" width="272" height="187">

```
print(s.hermite(0))
print(s.hermite(1))
print(s.hermite(2))
print(s.hermite(3))
```

```
# getting the coefficients
s.hermite(2)[2]
```

```
# value at point
s.hermite(2)(2)
```

```
for n in range(5):
    x = np.arange(-3,3,0.1)
    Hn = s.hermite(n)
    y = Hn(x)
    plt.plot(x,y)

plt.grid()
plt.ylim(-100,100)
plt.show()
```

Jn(x) and Yn(x) are the solutions.
Jn(n,x) is the format

```
s.jn(1,1)
```

```
s.yn(2,3)
```

```
x = np.linspace(0,10,200)
plt.plot(x,s.jn(1,x),x,s.jn(2,x),x,s.yn(1,x),x,s.yn(2,x))
plt.legend(['Jn(1)','Jn(2)','Yn(1)','Yn(2)'])
plt.grid()
plt.ylim(-1,1)
plt.show()
```
